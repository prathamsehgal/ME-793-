```python
# in this we can use nump.matrix,  scipy has the linalg.inv() function that computes the inverse of a matrix.
l =[5,10,20,50,100]
```

```python
X=[]
inverse=  []
table_1 = []

for i in l: 
    begin = time.time()
    print("MAtrix of dimension", i)               
    matrix = np.random.randint(10,size=(i,i))             #random matrix generated
    print(matrix)
    X.append(matrix)
    detX = np.linalg.det(matrix)                      #printing  determinant of X
  
    if detX!=0:                           # Checking if det is not 0 otherwise for singular matrix inv is not defined  
        invX=np.linalg.inv(matrix)                      #inverse of matrix calculated 
        print("the inverse for dimension ", i,"is", invX)

    
    else:
        print("Singular Matrix, Inverse doesn't exist for singular matrix")
        
    end = time.time()
    computed_time= end-begin
    print('computed time is ', computed_time)
    data = (i,computed_time*1000000)
    table_1.append(data)
        
```

```python
table_array= np.array(table_1)  # converting it into numpy array 
```

```python
df= pd.DataFrame(table_array, columns=['n','time in microseconds']) # making the dataframe for better view 
```
